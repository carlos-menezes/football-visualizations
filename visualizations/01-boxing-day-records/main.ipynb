```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create dataframe from the data
data = '''
Man Utd,28,22,4,2,78.6
Liverpool,26,17,5,4,65.4
Arsenal,25,16,7,2,64.0
Coventry,8,5,2,1,62.5
Spurs,26,15,8,3,57.7
Man City,21,12,2,7,57.1
Birmingham,6,3,2,1,50.0
Wigan,8,4,1,3,50.0
Chelsea,28,13,9,6,46.4
Stoke,9,4,2,3,44.4
Portsmouth,7,3,2,2,42.9
Wimbledon,7,3,2,2,42.9
Blackburn,17,7,8,2,41.2
Brighton,5,2,1,2,40.0
Leeds,10,4,2,4,40.0
Nottingham Forest,5,2,1,2,40.0
Everton,28,11,7,10,39.3
Swansea,6,2,1,3,33.3
West Ham,21,7,7,7,33.3
Sunderland,16,5,4,7,31.3
Southampton,20,6,7,7,30.0
Sheff Wed,7,2,3,2,28.6
Bolton,12,3,3,6,25.0
Crystal Palace,13,3,5,5,23.1
Middlesbrough,15,3,3,9,20.0
Newcastle,25,5,4,16,20.0
Wolves,5,1,1,3,20.0
Aston Villa,26,5,6,15,19.2
Fulham,16,3,8,5,18.8
Bournemouth,6,1,3,2,16.7
Charlton,7,1,3,3,14.3
Watford,7,1,3,3,14.3
Burnley,7,1,2,4,14.3
Derby,7,1,2,4,14.3
Leicester,17,2,3,12,11.8
West Brom,12,1,3,8,8.3
Norwich,8,0,1,7,0.0
Sheff Utd,6,0,2,4,0.0'''

# Convert the data into a DataFrame
df = pd.DataFrame([x.split(',') for x in data.split('\n')[1:]], 
                 columns=['Team', 'P', 'W', 'D', 'L', 'Win %'])

# Convert numeric columns
numeric_cols = ['P', 'W', 'D', 'L', 'Win %']
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric)

# Sort by win percentage
df = df.sort_values('Win %', ascending=True)

# Create the visualization
plt.figure(figsize=(12, 14))  # Increased height from 12 to 14
sns.set_style("white")

bars = plt.barh(df['Team'], df['Win %'], color="#d42220")

plt.title('Premier League Boxing Day Records\n', 
          fontsize=24,
          pad=20,
          fontweight=800,
          loc='left',
          family='SF Pro Display', color="#232322")

# Add subtitle
plt.text(0, plt.gca().get_ylim()[1] + 1,
         'Win percentage for all Premier League teams on December 26th',
         fontsize=14,
         color='#9a9a9b',
         family='SF Pro Text')
plt.text(80, -6,
         'Source: FBRef\nx: @c_mnzs\nbsky: @carlos-menezes',
         fontsize=10,
         color='#232322',
         ha="right",
         family='SF Pro Text')
plt.xlabel('Win %', fontsize=12, labelpad=15, color='#232322', family="SF Pro Text", fontweight=700)

# Remove the frame
plt.box(False)

# Customize grid
plt.grid(axis='x', linestyle='-.', alpha=0.3)
plt.grid(axis='y', linestyle='-.', alpha=0.3)

# Remove y-axis lines
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)

plt.gca().tick_params(axis='x', labelsize=10, labelfontfamily='SF Pro Text')
plt.gca().tick_params(axis='y', labelsize=10, labelfontfamily='SF Pro Text')

# Add value labels on the bars
for i, bar in enumerate(bars):
    width = bar.get_width()
    plt.text(width + 1, bar.get_y() + bar.get_height()/2,
             f'{width:.0f}%',
             ha='left', va='center', color='#232322', fontsize=10, family="SF Pro Text")


# Show the plot
plt.tight_layout()
plt.show()
```
